# isin() Function

```python
mask=delivery["over"]>15
death_over=delivery[mask]
```

```python
all_batsman=death_over.groupby("batsman")["batsman_runs"].count()
```

```python
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()                     #Boolean indexing to get all true values and tolist convert series to list
```

```python
#Runs scored by these batsman
#Balls play by these Batsman
```

```python
final=delivery[delivery["batsman"].isin(batsman_list)]
```

```python
runs=final.groupby("batsman")["batsman_runs"].sum()
```

```python
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
sr=(runs/balls)*100
```
