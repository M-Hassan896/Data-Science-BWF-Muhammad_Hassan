```python
#Converters Parameter
def rename(name):
    if name=="RCB":
        return "Royal Challengers Bangalore"
    else:
        return name
pd.read_csv("ipl_data.csv",converters={"team1":rename,"team2":rename})
```

```python
#na_values Parameter
pd.read_csv("ipl_data.csv",na_values=["RCB","MI"]) #replace RCB and MI with NaN
```
